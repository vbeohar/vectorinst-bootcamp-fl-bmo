###  Expanded correlation matrix with more fields 

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Convert timestamp to datetime if not already done
df['timestamp'] = pd.to_datetime(df['timestamp'], errors='coerce')

# Create fx_rate
df['fx_rate'] = np.where(df['amount_received'] == 0, np.nan, df['amount_paid'] / df['amount_received'])

# Time-based features
df['tx_hour'] = df['timestamp'].dt.hour
df['tx_day'] = df['timestamp'].dt.dayofweek
df['tx_month'] = df['timestamp'].dt.month

# Frequency-based features
df['account_pair'] = df['from_account'].astype(str) + ' → ' + df['to_account'].astype(str)
df['account_pair_freq'] = df.groupby('account_pair')['account_pair'].transform('count')
df['from_account_tx_count'] = df.groupby('from_account')['from_account'].transform('count')
df['to_account_rx_count'] = df.groupby('to_account')['to_account'].transform('count')

# Label encode from_bank, to_bank, and payment_format
label_cols = ['from_bank', 'to_bank', 'payment_format']
le = LabelEncoder()

for col in label_cols:
    df[col + '_encoded'] = le.fit_transform(df[col].astype(str))

# Select numeric features
numeric_df = df[[
    'amount_paid', 'amount_received', 'fx_rate',
    'tx_hour', 'tx_day', 'tx_month',
    'account_pair_freq', 'from_account_tx_count', 'to_account_rx_count',
    'from_bank_encoded', 'to_bank_encoded', 'payment_format_encoded',
    'is_laundering'
]]

# Drop rows with missing values
corr_df = numeric_df.dropna()

# Compute and plot correlation matrix
plt.figure(figsize=(14, 10))
sns.heatmap(corr_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Including Bank & Payment Format')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# # Load and rename columns
# df = pd.read_csv("HI-Medium_Trans.csv")

df.rename(columns={
    "From Bank": "from_bank",
    "To Bank": "to_bank",
    "Amount Received": "amount_received"
}, inplace=True)

# Combine from_bank and to_bank to get top 6 banks involved
all_banks = pd.concat([df['from_bank'], df['to_bank']])
top_banks = all_banks.value_counts().head(6).index  # Adjust number of banks here

# Subplot layout: 2 rows x 3 columns
fig, axes = plt.subplots(2, 3, figsize=(18, 8))
axes = axes.flatten()

# Plot one KDE per bank
for idx, bank in enumerate(top_banks):
    bank_df = df[(df['from_bank'] == bank) | (df['to_bank'] == bank)]
    filtered = bank_df[bank_df['amount_received'] > 0]

    sns.kdeplot(
        data=filtered,
        x="amount_received",
        fill=True,
        ax=axes[idx],
        alpha=0.6,
        linewidth=2
    )
    axes[idx].set_title(f"Bank {bank}")
    axes[idx].set_xlabel("Amount Received")
    axes[idx].set_ylabel("Density")
    axes[idx].grid(True, linestyle='--', alpha=0.3)

# Remove empty subplot if number of banks < total subplots
for j in range(len(top_banks), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

### Using wasserstein distance to check data distribution

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import wasserstein_distance
from tqdm import tqdm

# Load the dataset
df = pd.read_csv("../fl/data/HI-Medium_Trans.csv", low_memory=False)

# Rename columns for consistency
df.rename(columns={
    "Timestamp": "timestamp",
    "From Bank": "from_bank",
    "Account": "from_account",
    "To Bank": "to_bank",
    "Account.1": "to_account",
    "Amount Received": "amount_received",
    "Receiving Currency": "receiving_currency",
    "Amount Paid": "amount_paid",
    "Payment Currency": "payment_currency",
    "Payment Format": "payment_format",
    "Is Laundering": "is_laundering"
}, inplace=True)

# Ensure amount_received is numeric
df['amount_received'] = pd.to_numeric(df['amount_received'], errors='coerce')
```

```python
# Identify all unique bank IDs from both columns
unique_banks = pd.unique(df[['from_bank', 'to_bank']].values.ravel())

# Compute distributions for each bank (filter on either from_bank or to_bank)
bank_distributions = {}
for bank in tqdm(unique_banks, desc="Collecting bank distributions"):
    bank_df = df[(df['from_bank'] == bank) | (df['to_bank'] == bank)]
    bank_distributions[bank] = bank_df['amount_received'].dropna().values

# Compute pairwise Wasserstein distance (Earth Mover's Distance)
bank_pairs = []
distances = []

banks = list(bank_distributions.keys())
for i in range(len(banks)):
    for j in range(i + 1, len(banks)):
        b1, b2 = banks[i], banks[j]
        dist = wasserstein_distance(bank_distributions[b1], bank_distributions[b2])
        bank_pairs.append((b1, b2))
        distances.append(dist)
```

```python

# Create dataframe of dissimilarities
distance_df = pd.DataFrame(bank_pairs, columns=['bank1', 'bank2'])
distance_df['distance'] = distances

# Sort by most dissimilar
most_dissimilar = distance_df.sort_values(by='distance', ascending=False).head(10)

print("Top 10 most dissimilar bank pairs (by amount_received distribution):")
print(most_dissimilar)

# Optional: visualize KDE for the top 2 most dissimilar banks
top_b1, top_b2 = most_dissimilar.iloc[0][['bank1', 'bank2']]

plt.figure(figsize=(10, 5))
plt.title(f"Distribution of amount_received: Bank {top_b1} vs Bank {top_b2}")
sns.kdeplot(bank_distributions[top_b1], label=f'Bank {top_b1}')
sns.kdeplot(bank_distributions[top_b2], label=f'Bank {top_b2}')
plt.xlabel("Amount Received")
plt.ylabel("Density")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```
